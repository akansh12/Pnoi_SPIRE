```python
import pandas as pd
import numpy as np
```

```python
eko_Dataframe = pd.read_csv('./eko.csv')
```

```python
np.unique(eko_Dataframe.iloc[:,73])
Asthma = []
Post = []
Pre = []
Healthy = []
Male = []
Female = [] 
for i in np.unique(eko_Dataframe.iloc[:,73]):
        if i.find('sthma') !=-1:
            Asthma.append(i)
        else:
            Healthy.append(i)
        if i.find('post') !=-1:
            Post.append(i)
        if i.find('pre') != -1:
            Pre.append(i)
        if i.find('_M_') != -1:
            Male.append(i)
        if i.find('_F_')!= -1:
            Female.append(i)
Female_Asthma = []
Male_Asthma = []
for i in Asthma:
    if i.find('_F_') != -1:
        Female_Asthma.append(i)
    else:
        Male_Asthma.append(i)
```

```python
print(f'Asthma: {len(Asthma)}')
print(f'Healthy: {len(Healthy)}')
print(f'Male: {len(Male)}')
print(f'Female: {len(Female)}')
print(f'Pre: {len(Pre)}')
print(f'Post: {len(Post)}')
print(f'Female_Asthma: {len(Female_Asthma)}')
print(f'Male_Asthma: {len(Male_Asthma)}')



```
